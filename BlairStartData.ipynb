```python
#Import skater data csv file

import pandas as pd

NewDF = pd.read_csv("data/skaters_SS1.csv")
NewDF
```

```python
#Adds Plus/Minus estimates based on predicted team standings.

PlusMinusDF = pd.read_csv("data/PlusMinus.csv")

NewPlusMinusDF = pd.merge(NewDF, PlusMinusDF, on="TEAM", how="left")
NewPlusMinusDF
```

```python
NewPlusMinusDF.columns
```

```python
NewPlusMinusDF["Adj_PR"] = NewPlusMinusDF["PR"] + NewPlusMinusDF["PlusMinus"]*0.75
NewPlusMinusDF
```

```python
#Translates average auction prices into prices directly comparable to league.

AuctionPriceDF = pd.DataFrame(NewPlusMinusDF[['PLAYER', 'TEAM', 'NHLPOS' , 'AUC', 'Adj_PR']])

# Sets Max Avg alary in auction to expected max salary for league auction
MaxSalary = 60
MaxAuc = AuctionPriceDF['AUC'].max()
AuctionPriceDF['AUC180'] = AuctionPriceDF['AUC'] / MaxAuc * MaxSalary
AuctionPriceDF

# AuctionDollars = AuctionPriceDF['AUC'].sum()
# print(AuctionDollars)

AucDollars = AuctionPriceDF["AUC"].sum()
AuctionPriceDF['AUC180_2'] = (AuctionPriceDF['AUC'] / AucDollars) * (12 * 180 * 0.69)
AuctionPriceDF
```

```python
PRlistDF = pd.DataFrame(NewPlusMinusDF[['PLAYER', 'TEAM', 'NHLPOS' , 'Adj_PR']]).sort_values(by='Adj_PR', ascending=False)

PosList = []

for pos in PRlistDF['NHLPOS']:
    if pos == "R": 
        NewPos = "W"
    elif pos == "L": 
        NewPos = "W"
    else:
        NewPos = pos
        
    PosList.append(NewPos)

PRlistDF["Lg_Pos"] = PosList
        


#PRlistDF['PR_Share'] = PRlistDF['Adj_PR'] / PRlistDF['Adj_PR'].sum()
PRlistDF



#Nextsteps 
#1. Find marginal PR by pos
#2. Calc Marginal PR
#3. Calc Share of PR
#4. Calc Bid = PR share * skaterdollars
```

```python
#Nextsteps 
#1. Find marginal PR by pos
#2. Calc Marginal PR
#3. Calc Share of PR
#4. Calc Bid = PR share * skaterdollars



MarginalD_PR = PRlistDF.loc[PRlistDF['Lg_Pos']=="D"]["Adj_PR"].iloc[24]
print(MarginalD_PR)

MarginalW_PR = PRlistDF.loc[PRlistDF['Lg_Pos']=="W"]["Adj_PR"].iloc[36]
print(MarginalW_PR)

MarginalC_PR = PRlistDF.loc[PRlistDF['Lg_Pos']=="C"]["Adj_PR"].iloc[36]
MarginalC_PR

MarginCol = []

# #for pos in PRlistDF["Lg_Pos"]:
# for i = range(0, PRlistDF["PLAYER"].count)
#     TempRecord = PRlistDF.iloc[i]
#     if TempRecord['Lg_Pos'] == "W":
#         TempRecord['Marg_Val'] = max(0, TempRecord["Adj_PR"]-MarginalW_PR)
#     if TempRecord['Lg_Pos'] == "C":
#         TempRecord['Marg_Val'] = max(0, TempRecord["Adj_PR"]-MarginalC_PR)
#     if TempRecord['Lg_Pos'] == "D":
#         TempRecord['Marg_Val'] = max(0, TempRecord["Adj_PR"]-MarginalD_PR)
    
PRlistDF['Marg_Val'] = 0
    
    
for i in range(PRlistDF.shape[0]):
#     TempRecord = PRlistDF.iloc[i]
    if PRlistDF['Lg_Pos'][i] == "W":
         PRlistDF['Marg_Val'][i] = max(0, PRlistDF["Adj_PR"][i]-MarginalW_PR)
    if PRlistDF['Lg_Pos'][i] == "C":
         PRlistDF['Marg_Val'][i] = max(0, PRlistDF["Adj_PR"][i]-MarginalC_PR)
    if PRlistDF['Lg_Pos'][i] == "D":
         PRlistDF['Marg_Val'][i] = max(0, PRlistDF["Adj_PR"][i]-MarginalD_PR)

PRlistDF
```

```python
#3. Calc Share of PR

PR_Share = PRlistDF['Marg_Val'] / PRlistDF['Marg_Val'].sum()
PR_Share

#4. Calc Bid = PR share * skaterdollars
SalaryCap = 180
Teams = 12
TotalDollars = SalaryCap * Teams
GoalieDollars = 680
SkaterDollars = TotalDollars - GoalieDollars

PRlistDF["PlayerBid"] = PR_Share * SkaterDollars
PRlistDF
```

```python
print(AuctionPriceDF.head())
print(PRlistDF.head())

AvgAuctionDF = PRlistDF.merge(AuctionPriceDF, on="PLAYER", how='inner')
AvgAuctionDF

AuctionGraphDF = AvgAuctionDF.drop(columns=["NHLPOS_x", "NHLPOS_y", "Adj_PR_x", "Adj_PR_y", "TEAM_y", "AUC", "AUC180_2"])
AuctionGraphDF = AuctionGraphDF.sort_values(["PlayerBid"], ascending=False)
AuctionGraphDF = AuctionGraphDF.reset_index(drop=True,)
AuctionGraphDF = AuctionGraphDF.loc[AuctionGraphDF["PlayerBid"]>=1]

AuctionGraphDF["Bargain"] = AuctionGraphDF["PlayerBid"] - AuctionGraphDF["AUC180"]

AuctionGraphDF
```

```python


Goalie0DF = pd.read_csv("data/goalies_SS1.csv")

Goalie0DF.columns


```

```python
GoalieDF = pd.DataFrame(Goalie0DF[['PLAYER', 'TEAM', 'AUC', 'PR', 'P_GS', 'P_W','P_GAA','P_SVPCT', 'WFSI', 'GAAFSI', 'SVPFSI']]).sort_values(['PR'], ascending=False)
MarginalG = GoalieDF['PR'][23]

GoalieDF["Marg_Val"] = 0

for i in range(GoalieDF.shape[0]):
    GoalieDF["Marg_Val"][i] = max(0, GoalieDF['PR'][i] - MarginalG)

GoalieDF["PlayerBid"] = (GoalieDF["Marg_Val"]/GoalieDF["Marg_Val"].sum())*GoalieDollars
   
GoalieDF["AUC180"] = GoalieDF["AUC"]/GoalieDF["AUC"].max()*50
GoalieDF = GoalieDF.reset_index(drop=True,)

GoalieDF["Lg_Pos"] = "G"

GoalieDF

StartGoaliesDF = GoalieDF.iloc[:24]
StartGoaliesDF

#StartGoaliesDF['P_GAA'].median
#StartGoaliesDF['P_SVPCT'].median
```
